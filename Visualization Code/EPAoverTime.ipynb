```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import nflfastpy as nfl
!pip install pathlib
```

```python
data = pd.concat([nfl.load_pbp_data(season).assign(season=season) for season in range(1999, 2021)])
```

```python
passing_offense_epa = data.loc[data['pass_attempt'] == 1, :]\
.groupby(['season'], as_index=False)['epa'].mean()
```

```python
df = passing_offense_epa
df

from pathlib import Path  
filepath = Path('/Users/moore_austin02/Documents/School Files/EpaOverTime.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
df.to_csv(filepath)  
```

```python
x = df.season
y = df.epa
z = np.sqrt(x**2 + y**2)
img = plt.imread('NFL1.png')




fig, ax = plt.subplots(figsize = (12,7))

ax.scatter(x, y, c='green', s=200, marker=(4, 2))

plt.style.use('fivethirtyeight')
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300



#Add a grid
ax.grid(zorder=1,alpha=.8)
ax.set_axisbelow(True)

    
#Adding labels and text\
ax.set_title('EPA/Pass Attempt Over Time', fontsize=20)
ax.set_xlabel('Season', fontsize=14)
ax.set_ylabel('EPA per Play', fontsize=14)
plt.figtext(.78, .0, 'Data from nflfastR @austinmoore18', fontsize=10)
#plt.savefig('Early Down Run Rate by EPA per Play4.png')
```
